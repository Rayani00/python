## Matplotlib example workflow

```python
# 0. Immport matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare the data
x = [1,2,3,4]
y = [25,22,13,65]

# 2. Setup plot
fig, ax = plt.subplots(figsize= (10,10)) # (width, height)

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show the figure
fig.savefig("sample-plot.png")
```

 ### Two options for subplots

```python
# Subplot option 1 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.rand(10),np.random.rand(10));
ax3.hist(np.random.randint(0,10,5));
ax4.bar(fruits_prices.keys(),fruits_prices.values());
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(fruits_prices.keys(),fruits_prices.values());
ax[1,1].hist(np.random.randint(0,10,10))
```

### Which one should you use ? (pyplot or matplotlib OO method)

* When ploting something quickly ok to use pyplot
* When ploting some more advanced, use the OO method

```python
# Lets grab the people that have age > 50
over50 = heartDisease[heartDisease["age"] > 50]
over50
```

```python
over50.age.mean()
```

```python
# pyplot method
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target');
```

```python
# OO method
fig, ax=plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig, ax= plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target'],
                    )

# Customize the plot
ax.set(title='Heart disease and cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");

# Add a horizontal line for the mean of chol level
ax.axhline(over50["chol"].mean(),
          linestyle='--');
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")

# subplot of chol, age, thalach 
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,6),
                             sharex=True) # this changes the colour sheme;

# Plot the scatters
ax0.scatter(x=over50['age'],
           y=over50['chol'],
           c=over50['target'],
           cmap="winter");
# Remove the lim
ax0.set_xlim([50,80])
ax0.set_ylim([0,600])

ax1.scatter(x=over50['age'],
           y=over50['thalach'],
           c=over50['target'],
           cmap="plasma");

# Beautify our scatters
ax1.set(title='Heart diseases and thalach levels.',
       xlabel='Age',
       ylabel='thalach');

ax0.set(title='Heart diseases and cholesterol levels.',
       ylabel='Cholesterol');

#Add legends to the scatters
ax0.legend(*scatter.legend_elements(),title='Target');
ax1.legend(*scatter.legend_elements(),title='Target');

# Add the horizontal line of mean
ax0.axhline(over50['chol'].mean(),linestyle='--')
ax1.axhline(over50['thalach'].mean(),linestyle='--')

# Add a title to the figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold')
```
