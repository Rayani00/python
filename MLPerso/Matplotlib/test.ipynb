```python
# 0. Immport matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
# 1. Prepare the data
x = [1,2,3,4]
y = [10,33,45,60]

# 2. Setup plot
fig, ax = plt.subplots(figsize= (10,10)) # (width, height)

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="L'effet du digital sur la triche",
      xlabel="Digitalisation",
      ylabel="Triche")

# 5. Save & show the figure
fig.savefig("triche-digital.png")
```
